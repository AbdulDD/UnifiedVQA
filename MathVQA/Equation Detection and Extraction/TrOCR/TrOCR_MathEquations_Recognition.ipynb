```python
import torch
from torch.utils.data import Dataset
from PIL import Image

class IAMDataset(Dataset):
    def __init__(self, root_dir, df, processor, max_target_length=256):
        self.root_dir = root_dir
        self.df = df
        self.processor = processor
        self.max_target_length = max_target_length

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        # get file name + text
        file_name = self.df['filename'][idx]
        text = self.df['text'][idx]
        # prepare image (i.e. resize + normalize)
        image = Image.open(self.root_dir + file_name).convert("RGB")
        pixel_values = self.processor(image, return_tensors="pt").pixel_values
        # add labels (input_ids) by encoding the text
        labels = self.processor.tokenizer(text,
                                          padding="max_length",
                                          max_length=self.max_target_length).input_ids
        # important: make sure that PAD tokens are ignored by the loss function
        labels = [label if label != self.processor.tokenizer.pad_token_id else -100 for label in labels]

        encoding = {"pixel_values": pixel_values.squeeze(), "labels": torch.tensor(labels)}
        return encoding
```

## Model Token setup

```python
# set special tokens used for creating the decoder_input_ids from the labels
model.config.decoder_start_token_id = processor.tokenizer.cls_token_id
model.config.pad_token_id = processor.tokenizer.pad_token_id
# make sure vocab size is set correctly
model.config.vocab_size = model.config.decoder.vocab_size

# set beam search parameters
model.config.eos_token_id = processor.tokenizer.sep_token_id
model.config.max_length = 128
model.config.early_stopping = True
model.config.no_repeat_ngram_size = 3
model.config.length_penalty = 2.0
model.config.num_beams = 4
```

## Evaluation metric

```python
from datasets import load_metric

cer_metric = load_metric("cer")

def compute_cer(pred_ids, label_ids):
    pred_str = processor.batch_decode(pred_ids, skip_special_tokens=True)
    label_ids[label_ids == -100] = processor.tokenizer.pad_token_id
    label_str = processor.batch_decode(label_ids, skip_special_tokens=True)

    cer = cer_metric.compute(predictions=pred_str, references=label_str)

    return cer
```

```python
# from transformers import AdamW
# from tqdm.notebook import tqdm
# import torch
# import os

# # Define optimizer with weight decay
# optimizer = AdamW(model.parameters(), lr=1e-5)

# # Learning rate scheduler
# #scheduler = torch.optim.lr_scheduler.StepLR(optimizer, step_size=5, gamma=0.5)

# # Gradient clipping value
# max_grad_norm = 1.0

# # Base save directory
# base_save_directory = r"E:\Abdul_Muqtadir\Thesis\OCR\TrOCR\save_directory"

# # Early stopping threshold
# best_cer = float('inf')
# patience = 5
# patience_counter = 0

# for epoch in range(20):
#     # Training
#     model.train()
#     train_loss = 0.0
#     for batch in tqdm(train_dataloader):
#         # Move batch to device
#         for k, v in batch.items():
#             #print(k, v)
#             batch[k] = v.to(device)

#         # Forward + backward + optimize
#         outputs = model(**batch)
#         loss = outputs.loss
#         loss.backward()

#         # Clip gradients
#         torch.nn.utils.clip_grad_norm_(model.parameters(), max_grad_norm)

#         optimizer.step()
#         optimizer.zero_grad()

#         train_loss += loss.item()

#     # Scheduler step
#     scheduler.step()

#     print(f"Loss after epoch {epoch}: {train_loss / len(train_dataloader)}")

#     # Evaluation
#     model.eval()
#     valid_cer = 0.0
#     with torch.no_grad():
#         for batch in tqdm(eval_dataloader):
#             # Generate outputs
#             outputs = model.generate(
#                 batch["pixel_values"].to(device),
#                 max_length=128,
#                 early_stopping=True,
#                 num_beams=4,
#                 length_penalty=2.0,
#                 no_repeat_ngram_size=3
#             )
#             # Compute CER
#             cer = compute_cer(pred_ids=outputs, label_ids=batch["labels"])
#             valid_cer += cer

#     avg_valid_cer = valid_cer / len(eval_dataloader)
#     print("Validation CER:", avg_valid_cer)

#     # Save model if CER improves
#     if avg_valid_cer < best_cer:
#         best_cer = avg_valid_cer
#         patience_counter = 0
#         epoch_save_directory = os.path.join(base_save_directory, f"epoch_{epoch + 1}")
#         os.makedirs(epoch_save_directory, exist_ok=True)
#         model.save_pretrained(epoch_save_directory)
#         print(f"Model saved to {epoch_save_directory}")
#     else:
#         patience_counter += 1
#         print(f"No improvement in CER. Patience: {patience_counter}/{patience}")

#     # Stop training if patience is exceeded
#     if patience_counter >= patience:
#         print("Early stopping triggered.")
#         break
from transformers import AdamW
from tqdm.notebook import tqdm
import torch
import os

# Define optimizer
optimizer = AdamW(model.parameters(), lr=5e-5)

# Gradient clipping value
max_grad_norm = 1.0

# Directory for saving models
output_dir = r"E:\Abdul_Muqtadir\Thesis\OCR\TrOCR\save_directory"

# Best CER and patience settings
best_cer = float('inf')
patience = 5
patience_counter = 0

for epoch in range(10):  # Loop over dataset multiple times
    # Training
    model.train()
    train_loss = 0.0
    for batch in tqdm(train_dataloader, desc=f"Epoch {epoch + 1} Training"):
        # Move batch to device
        for k, v in batch.items():
            batch[k] = v.to(device)

        # Forward + backward + optimize
        outputs = model(**batch)
        loss = outputs.loss
        loss.backward()

        # Clip gradients
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_grad_norm)

        optimizer.step()
        optimizer.zero_grad()

        train_loss += loss.item()

    print(f"Epoch {epoch + 1} Training Loss: {train_loss / len(train_dataloader):.4f}")

    # Evaluation
    model.eval()
    valid_cer = 0.0
    with torch.no_grad():
        for batch in tqdm(eval_dataloader, desc=f"Epoch {epoch + 1} Evaluation"):
            # Generate outputs
            outputs = model.generate(
                batch["pixel_values"].to(device),
                max_length=128,
                early_stopping=True,
                num_beams=4,
                length_penalty=2.0,
                no_repeat_ngram_size=3
            )
            # Compute CER
            cer = compute_cer(pred_ids=outputs, label_ids=batch["labels"])
            valid_cer += cer

    avg_valid_cer = valid_cer / len(eval_dataloader)
    print(f"Epoch {epoch + 1} Validation CER: {avg_valid_cer:.4f}")

    # Save model if CER improves
    if avg_valid_cer < best_cer:
        best_cer = avg_valid_cer
        patience_counter = 0
        epoch_save_directory = os.path.join(output_dir, f"epoch_{epoch + 1}")
        os.makedirs(epoch_save_directory, exist_ok=True)
        model.save_pretrained(epoch_save_directory)
        print(f"Model saved to {epoch_save_directory}")
    else:
        patience_counter += 1
        print(f"No improvement in CER. Patience: {patience_counter}/{patience}")

    # Early stopping
    if patience_counter >= patience:
        print("Early stopping triggered.")
        break

```
